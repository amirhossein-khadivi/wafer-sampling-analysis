# Case Study

```python
from wafer_utils import create_wafer_mask, plot_wafer_mask

wafer_size = 33
mask, center = create_wafer_mask(wafer_size=wafer_size)
plot_wafer_mask(mask, center)
```

# Sample Size (Prop Method)

```python
from sample_size_calc import calculate_sample_size

sample_size = calculate_sample_size(confidence_level=0.95, p=0.2, E=0.05)
print(f"Required sample size: {sample_size}")
```

# Sampling

## Simple Random Sampling

```python
from srs import sample_and_plot

samples, valid_coords = sample_and_plot(mask=mask, center=center, sample_size=sample_size, wafer_size=wafer_size)
```

## Systematic Sampling

```python
from sys_sample import systematic_sampling, plot_systematic_samples

samples = systematic_sampling(mask, sample_size, valid_coords)
plot_systematic_samples(mask, center, samples)
```

## Stratified Sampling

```python
from str_sample import stratified_sampling, plot_stratified_samples

sample_stratified, strata_indices, valid_coords = stratified_sampling(mask, center, sample_size, valid_coords, n_strata=3)
plot_stratified_samples(mask, center, sample_stratified, strata_indices, valid_coords, n_strata=3)
```
